# Halite Map Blurring
---

### Simulate the first steps of a game to obtain a nice halite map

```python
from kaggle_environments import make
from haliteivbot.rule_based.bot import agent
```

```python
agent_count = 4
steps = 10
env = make("halite", configuration={"size": 21, "startingHalite": 5000, "episodeSteps": 100}, debug=True)
info = env.reset(agent_count)
```

```python
_ = env.run(["random", "random", "random", agent])
```

### Plot the map

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

halite_map = np.array(env.state[0].observation.halite).reshape((21, -1))
plt.figure(figsize=(10, 10))
plt.imshow(halite_map, cmap='plasma')
```

### Try different blurs
---

```python
from scipy.ndimage import gaussian_filter
# Gaussian blur
for sigma in np.arange(0.25, 1.25, 0.25):
    blurred_map = gaussian_filter(halite_map, sigma, mode='wrap')
    plt.figure(figsize=(5, 5))
    plt.title("Gaussian blur (sigma = {})".format(sigma))
    plt.imshow(blurred_map, cmap='plasma')
```
